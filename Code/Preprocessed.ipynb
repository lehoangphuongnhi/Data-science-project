## Tiền xử lý dữ liệu
- Kiểm tra tập dữ liệu có dữ liệu trùng
- Thuộc tính `display colors`:
    - Có các giá trị *Type IPS L, Type OLED ,Type P-OLE,  Type AMOLE, Type Super, Type LTPS* là không đúng với tính chất của thuộc tính
    - Chuyển giá trị trong cột về cùng một đơn vị, và chuyển về dạng số
- Thuộc tính `screen size`, `screen area`, `screen to body ratio` chuyển về dạng số
- Thuộc tính `screen resolution` lấy chiều dài nhân chiều rộng để ra số pixel, chuyển `screen resolution` về dạng số
- Thuộc tính `rear camera`, `front camera` lấy giá trị là độ phân giải, chuyển thuộc tính về dạng số
- Thuộc tính `os` lấy tên hệ điều hành và số ghi thế hệ của hệ điều hành
- Thuộc tính `gpu` xóa khí tự "-"
- Thuộc tính `external memory`chia ra làm hai cột:
    - Một cột sẽ lưu là 1 nếu điện thoại có external memory, là 0 nếu điện thoại không có external memory
    - Một cột sẽ lưu là dung lượng của external memory, chuyển cột này về dạng số
- Thuộc tính `internal_memory` lấy giá trị của dung lượng RAM, chuyển cột về dạng số
- Thuộc tính `cpu` lấy giá trị xung nhịp, và chuyển các dòng về cùng một đơn vị, chuyển về dạng số
- Thuộc tính `dimensions` chia ra làm ba cột:
     - `length`: Chiều dài, sau đó chuyển cột về dạng số
     - `width`: Chiều rộng, sau đó chuyển cột về dạng số
     - `high`: Độ dày, sau đó chuyển cột về dạng số
- Thuộc tính `battery` chia ra làm ba cột: 
    - `battery_mah` lưu khả năng trữ điện của pin, đưa về dạng số
    - `battery_removable` bằng 1 khi có pin rời, bằng 0 khi pin liền
    - `battery_type` lưu loại pin
- Thuộc tính `approximate price` chuyển về cùng một đơn vị tiền tệ và chuyển về dạng số, sau đó sẽ tách làm 3 lớp {0, 1, 2} tương ứng với mức giá x < 150, 150 <= x < 300, x >= 300 (x là giá điện thoại)
- Thuộc tính `net_work` chia làm 4 cột `_2G`, `_3G`, `_4G`, `_5G`, các cột lưu giá trị {1, 0} thể hiện điện thoại có hỗ trợ loại mạng đó không
- Thuộc tính `radio`, `headphone_jack`, `gprs`, `gps`, `usb`, `wifi`, `bluetooth` thay giá trị 1 nếu điện thoại có thuộc tính, bằng 0 nếu điện thoại không có thuộc tính

```python
phones_df = phones_df.drop_duplicates()
```

```python
n = phones_df.display_colors.replace(to_replace = ['Type IPS L', 'Type OLED' ,'Type P-OLE',  'Type AMOLE', 'Type Super', 'Type LTPS'], value = 'None')
```

```python
phones_df.display_colors = pd.to_numeric(n.replace('256K',  0.256).replace('16M',  16).replace('16M',  16).replace('65K',  0.065)
              .replace('1B',  1000).replace('16М',  16),errors='coerce')
```

```python
phones_df.screen_size = pd.to_numeric(phones_df.screen_size.str.replace('" inc',''), errors='coerce')
```

```python
phones_df.screen_area = pd.to_numeric(phones_df.screen_area.str.replace("cm2",'').str.replace(',', '.'), errors='coerce')
```

```python
phones_df.screen_to_body_ratio = pd.to_numeric(phones_df.screen_to_body_ratio.str.replace("%",'').str.replace(',', '.'), errors='coerce')
```

```python
phones_df.screen_resolution = pd.to_numeric(phones_df.screen_resolution.str.extract(r'( [\d]+)', expand=False), errors='coerce')*pd.to_numeric(phones_df.screen_resolution.str.extract(r'([\d]+)', expand=False), errors='coerce')
```

```python
phones_df.rear_camera = pd.to_numeric(phones_df.rear_camera.str.extract(r'([\d.]+)', expand=False), errors='coerce')
```

```python
phones_df.front_camera = pd.to_numeric(phones_df.front_camera.str.extract(r'([\d.]+)', expand=False), errors='coerce')
```

```python
phones_df.os = phones_df.os.str.extract(r'([a-zA-z ]+ [\d.]+)', expand=False)
```

```python
phones_df.gpu = phones_df.gpu.str.replace('-', ' ').str.strip().str.extract(r'([a-zA-z\d]+ [a-zA-z\d]+|[a-zA-z\d]+)', expand=False)
```

```python
external_memory = phones_df.external_memory.replace('No', 0)
external_memory.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
#Hàm chuyển đổi đơn vị cho memory
def memory(x):
    if str(x) != 'nan':
        if x.find('GB') != -1:
            return float(re.findall('\d+', x)[0])
        elif x.find('TB') != -1:
            return float(re.findall('\d+', x)[0])*1024
        elif x.find('MB') != -1:
            return float(re.findall('\d+', x)[0])/1024
    else:
        return np.nan
```

```python
external_memory_size = phones_df.external_memory.str.extract(r'([\d.]+[a-zA-z ]+)', expand=False).apply(lambda x: memory(x))
```

```python
phones_df.external_memory = external_memory
phones_df['external_memory_size'] = external_memory_size
```

```python
phones_df.internal_memory = phones_df.internal_memory.str.extract(r'([\d ]+[a-zA-Z ]+RAM)', expand = False).str.extract(r'([\d ]+[a-zA-Z]+)',expand = False).apply(lambda x: memory(x))
```

```python
#Hàm chuyển đổi đơn vị cho cpu
def cpu(x):
    if str(x) != 'nan':
        if x.find('GHz') != -1:
            return float(re.findall('[\d.]+', x)[0])
        elif x.find('MHz') != -1:
            return float(re.findall('[\d.]+', x)[0])*0.001
    else:
        return np.nan
```

```python
phones_df.cpu = phones_df.cpu.str.extract(r'([\d.]+ GHz|[\d.]+ MHz|[\d.]+MHz|[\d.]+GHz)', expand=False).apply(lambda x: cpu(x))
```

```python
phones_df['length'] = pd.to_numeric(phones_df.dimensions.str.extract(r'([\d.]+)', expand=False), errors='coerce')
```

```python
phones_df['width'] = pd.to_numeric(phones_df.dimensions.str.extract(r'( [\d.]+)', expand=False), errors='coerce')
```

```python
phones_df['high'] = pd.to_numeric(phones_df.dimensions.str.extract(r'( [\d.]+ [a-zA-z])', expand=False).str.replace(' m',''), errors='coerce')
```

```python
phones_df.weight = pd.to_numeric(phones_df.weight.str.extract(r'([\d.]+)', expand=False).replace('0', np.nan), errors='coerce')
```

```python
def type(battery):
    if battery.find('Li-Ion') != -1:
        return 'Li-Ion'
    elif battery.find('Li-Po') != -1:
        return 'Li-Po'
    else:
        return np.nan
```

```python
phones_df['battery_mah'] = pd.to_numeric(phones_df.battery.str.extract(r'([\d.]+)', expand=False), errors='coerce')
phones_df['battery_removable'] = phones_df.battery.str.replace('\xa0', '').str.contains('Removable')
phones_df['battery_removable'].replace(True, 1, inplace=True)
phones_df['battery_removable'].replace(False, 0, inplace=True)
phones_df['battery_type'] = phones_df.battery.apply(lambda x: type(x))
```

```python
currency = ['EU', 'USD', 'INR', 'BTC', '€', '₹']
```

```python
phones_df['price'] = pd.to_numeric(phones_df.approximate_price.str.extract(r'([\d.]+)',expand=False), errors='coerce').copy()
```

```python
for i in  currency:
    phones_df[i] = phones_df.approximate_price.str.contains(i).copy().replace(True, 1).replace(False, 0) 
a = []
for i in  phones_df.index:
        j = '$'
        if j  in phones_df.approximate_price[i]:
            a.append(1)
        else:
            a.append(0)
```

```python
phones_df['real_price'] = phones_df.price * (phones_df.EU* 1 + phones_df.USD*0.82 + phones_df.INR* 0.011 + phones_df[currency[4]] * 1 + phones_df[currency[5]]*0.011)
```

```python
phones_df  = phones_df[phones_df['real_price'] != 0].copy()
```

```python
for i in phones_df.real_price:
    if i < 150:
        phones_df.real_price.replace(i, 0, inplace = True)
for i in phones_df.real_price:
    if (i >= 150 and i < 300):
        phones_df.real_price.replace(i, 1, inplace = True)
    elif (i >= 300):
        phones_df.real_price.replace(i, 2, inplace = True)
```

```python
phones_df['_2G'] = phones_df.net_work.str.contains('GSM').copy().replace(True, 1).replace(False, 0)
phones_df['_3G'] = phones_df.net_work.str.contains('CDMA2000| EVDO|UMTS|HSPA').copy().replace(True, 1).replace(False, 0)
phones_df['_4G'] = phones_df.net_work.str.contains('LTE').copy().replace(True, 1).replace(False, 0)
phones_df['_5G'] =  phones_df.net_work.str.contains('5G').copy().replace(True, 1).replace(False, 0)
```

```python
phones_df.radio.replace('No',  0, inplace = True)
phones_df.radio.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
phones_df.headphone_jack.replace(['No, included adaptor for 3.5mm', 'No'],value = 0, inplace = True)
phones_df.headphone_jack.replace(['Yes', 'TBD', '1'],value = 1, inplace = True)
```

```python
phones_df.gprs.replace('No', 0, inplace = True)
phones_df.gprs.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
phones_df.gps.replace('No', 0, inplace = True)
phones_df.gps.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
phones_df.wifi.replace(['No', np.nan], 0, inplace = True)
phones_df.wifi.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
phones_df.usb.replace(['No', np.nan], 0, inplace = True)
phones_df.usb.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```

```python
phones_df.bluetooth.replace(['No', np.nan], 0, inplace = True)
phones_df.bluetooth.replace(regex = r'([a-zA-Z]+)', value = 1, inplace = True)
```
