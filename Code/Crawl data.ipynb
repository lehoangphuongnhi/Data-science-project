```python
from requests_html import HTMLSession
import pandas as pd
import numpy
import urllib.robotparser
import time
from bs4 import BeautifulSoup
import json
import urllib3
urllib3.disable_warnings()
import requests
```

## Kiểm tra file Robots.txt của trang

```python
rp = urllib.robotparser.RobotFileParser()
rp.set_url('https://phonesdata.com/robots.txt')
rp.read()
rp.can_fetch('*', 'https://phonesdata.com/en/smartphones/')
```

---

## Lấy link các hãng điện thoại

```python
url = 'https://phonesdata.com/en/smartphones/'
session = HTMLSession()
res = session.get(url)
```

```python
links_brand = []
for html in res.html:
    list_brand = html.find('div.product-item')
    for brand in list_brand:
        url = brand.find('a', first = True).attrs['href']
        links_brand.append(url)
```

```python
len(links_brand)
```

---

## Lấy link điện thoại theo từng năm

```python
links_phone_year = []
for url in links_brand:
    html_text = requests.get(url).text
    tree = BeautifulSoup(html_text, 'html.parser')
    for link in tree.find_all('div', {'class': 'col-md-12 col-sm-12 col-xs-12'}):
        m = link.find_all('a')
        count = 0
        for phone in m:
            link_phone = phone.get('href')
            if ('#' in link_phone):
                if count == 0:
                    links_phone_year.append(link_phone)
                    print(link_phone)
                    count += 1
            else:
                links_phone_year.append(link_phone)
                print(link_phone)
```

```python
len(links_phone_year)
```

## Lấy link và tên hãng điện thoại

```python
links_phone = []
brand = []
for url in links_phone_year:
    html_text = requests.get(url).text
    tree = BeautifulSoup(html_text, 'html.parser')
    if (tree.find_all('div', {'class': 'pi-img-wrapper'}) != None):
        for link in tree.find_all('div', {'class': 'pi-img-wrapper'}):
            if (len(tree.find_all('span', {'property': 'name'})) > 2):
                brand.append(tree.find_all('span', {'property': 'name'})[2].text)
                test = tree.find_all('span', {'property': 'name'})[2].text
            else:
                brand.append('Fairphone')
                test = 'Fairphone'
            print(test)
            m = link.find_all('a')
            for phone in m:
                link_phone = phone.get('href')
                print(link_phone)
                links_phone.append(link_phone)
```
